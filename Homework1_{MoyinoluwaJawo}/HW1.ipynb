```python
import thinkstats2
import thinkplot
hist = thinkstats2.Hist(resp.fmarno, label='marriage')
thinkplot.Hist(hist,width=0.5)
thinkplot.Config(xlabel='Number of Marriages', ylabel='Frequency')
```

```python
neverMarried=resp[resp.fmarno==0]
married=resp[resp.fmarno>=1]

nevMarried_hist=thinkstats2.Hist(neverMarried.totincr, label='never Married')
married_hist=thinkstats2.Hist(married.totincr, label='Married')

width = 0.5
#thinkplot.PrePlot(2)
thinkplot.Hist(nevMarried_hist, align='left', width=width)
thinkplot.Hist(married_hist, align='right', width=width)
thinkplot.Config(xlabel='Total income', ylabel='No. of Respondents')
```

```python
min_inc1=neverMarried.totincr.min()
min_inc2=married.totincr.min()
print('The minimum income of non-married respondents is %d'%(min_inc1))
print('The minimum income of married respondents is %d'%(min_inc2))

print('The maximum total income of non-married respondents is %d'%(neverMarried.totincr.max()))
print('The maximum total income of married respondents is %d'%(married.totincr.max()))

print('The standard deviation of the income of non-married respondents is %f'%(neverMarried.totincr.std()))
print('The standard deviation of the income of married respondents is %f'%(married.totincr.std()))

print('The variance of the income of non-married respondents is %f'%(neverMarried.totincr.var()))
print('The variance of the income of married respondents is %f'%(married.totincr.var()))

print('The mean income of non-married respondents is %f'%(neverMarried.totincr.mean()))
print('The mean income of married respondents is %f'%(married.totincr.mean()))
```
